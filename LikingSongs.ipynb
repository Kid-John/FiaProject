# Data Cleaning

Rimozione di alcune colonne inutili

```python
#Rimozione colonne inutili
datasetSongs.drop(['added','top year', 'acous', 'spch', 'pop', 'dB','live','nrgy', 'dnce','val','bpm'], axis=1, inplace=True)

#Formattazione Nome delle Colonne
datasetSongs.columns = datasetSongs.columns.str.title()

datasetSongs
```

Formattazione Anno di Rilascio

```python
datasetSongs[['Year Released']] = datasetSongs[['Year Released']].apply(lambda x: x.astype(str).str.split('.').str[0])

datasetSongs
```

Formattazione Tipo di Artista

```python
datasetSongs['Artist Type'] = datasetSongs['Artist Type'].apply(lambda x: str(x).split('/')[0])

datasetSongs
```

Modifica del titolo della colonna Top Genre in Genre

```python
datasetSongs['Genre'] = datasetSongs['Top Genre']
datasetSongs.drop(['Top Genre'], axis=1, inplace=True)

datasetSongs['Genre'] = datasetSongs['Genre'].str.title()

datasetSongs
```

Elimina tutte le righe nulle

```python
datasetSongs.dropna(inplace=True)

datasetSongs
```

Controllo che le righe nulle siano 0 per ogni colonna

```python
nan_mask = datasetSongs.isna()
nan_count = nan_mask.sum()

nan_count
```

Formattazione e Rimozione di alcune colonne del dataset con i valori di gradibilità

```python
datasetSongsMe.drop(['bpm'], axis=1, inplace=True)

datasetSongsMe.columns = datasetSongsMe.columns.str.title()
datasetSongsMe['Top Genre'] = datasetSongsMe['Top Genre'].str.title()

datasetSongsMe
```

```python
datasetSongsMe.groupby(['Title'])

x = 0
for i, row in datasetSongsMe.iterrows():
  title = row['Title']
  if datasetSongs['Title'].eq(title).any():
    x += 1

print(x)
print(datasetSongsMe['Title'].count)

datasetSongsMe

```

Controllo dei dati per il merge successivo

```python
print("Tipi di dati in datasetMe:")
print(datasetSongsMe.dtypes)

print("\nTipi di dati in dataset:")
print(datasetSongs.dtypes)
```

Faccio il merge tra il dataset iniziale e quello con i valori di gradimento

```python
datasetSongs['Year Released'] = datasetSongs['Year Released'].astype(int)

df = pd.merge(datasetSongsMe, datasetSongs, left_on=['Title','Artist','Year Released','Dur','Artist Type','Top Genre'], right_on=['Title','Artist','Year Released','Dur','Artist Type','Genre'],how ='left')
df = df.drop_duplicates(subset=['Title','Artist'])

df.drop(['Top Genre'], axis=1, inplace=True)

df
```

```python
for i, row in df.iterrows():
  dur = row['Dur']
  if dur <= 240 :
    df.loc[i, 'Dur'] = 'Media'
  elif dur < 180 :
    df.loc[i, 'Dur'] = 'Corta'
  else:
    df.loc[i, 'Dur'] = "Lunga"

print(df['Dur'].value_counts())
df
```

Inserisco mi piace se il valore è 1 e non mi piace se è 0

```python
df['Liking'] = pd.to_numeric(df['Liking'], errors='coerce')
for i, row in df.iterrows():
  liking = row['Liking']
  if liking == 1 :
    df.loc[i, 'Liking'] = 'Mi Piace'
  else:
    df.loc[i, 'Liking'] = 'Non Mi Piace'

#Mostra il conteggio di Mi Piace e Non Mi Piace
print(df['Liking'].value_counts())
df
```

#Gestione dei valori nulli


Elimino i valori nulli se sono presenti

#Inizializzazione e Addestramento del Modello

Inizializzo e Addestro il modello secondo il dataset merge ottenuto in precedenza

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report,  precision_score
from sklearn.preprocessing import OneHotEncoder

# Selezione delle colonne del DataFrame utili per l'addestramento del modello
X = df[['Artist','Year Released', 'Genre', 'Dur','Artist Type']]
y = df[['Liking']]

# Utilizzo di OneHotEncoder per gestire la codifica di variabili categoriche
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X)


# Separiamo i nostri dati in dati di training e dati di test
# Specifichiamo la proporzione fra i due con test_size; in questo caso abbiamo impostato il training set al 90% e il test set al 10%
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.1, random_state=42)

#Dopo una prova decido di utilizzare l'albero decisionale poichè l'accuracy era quella più alta
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)


print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{classification_rep}")
```

Stampo i valori predetti dal modello cosi da poterli confrontare con quelli reali

```python
#Aggiungo una colonna all'interno del dataset merge con i dati predetti dal modello
df['Predicted_Liking'] = model.predict(X_encoded)

matching_values = df['Liking'] == df['Predicted_Liking']

#Conta il numero di righe che sono uguali a quelli scritti nel dataset con i valori di gradimento
matching_count = matching_values.sum()
print(f"Numero di righe con valori uguali: {matching_count}/241")

df
```
